# Dealing with missing Values

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
missing_values_percentage = (df.isnull().sum() / len(df)) * 100
print(missing_values_percentage)
```

```python
columns_with_high_missing_values = missing_values_percentage[missing_values_percentage > 60]
print(columns_with_high_missing_values)
```

```python
df = df.drop(['pf_identity_inheritance_widows', 'pf_identity_inheritance_daughters'], axis=1)
```

```python
df = df.fillna(df.mean())
```

# Checking for duplicates

```python
duplicated_records = df[df.duplicated()]

if duplicated_records.empty:
    print("No duplicated records found.")
else:
    print("Duplicated records:")
    print(duplicated_records)
```

# Address outliers

```python
import seaborn as sns
sns.boxplot(data=df)
```

```python
import numpy as np
import matplotlib.pyplot as plt
threshold = 6


threshold = 6

numeric_columns = df.select_dtypes(include=np.number).columns


z_scores = np.abs((df[numeric_columns] - df[numeric_columns].mean()) / df[numeric_columns].std())
columns_with_outliers = z_scores.columns[z_scores.max() > threshold]


num_plots = len(columns_with_outliers)
num_rows = int(np.sqrt(num_plots))
num_cols = int(np.ceil(num_plots / num_rows))


fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 8))


axes = axes.flatten()


for i, column in enumerate(columns_with_outliers):
    sns.boxplot(data=df[column], ax=axes[i])
    axes[i].set_title(column)

if len(columns_with_outliers) < num_plots:
    for j in range(len(columns_with_outliers), num_plots):
        fig.delaxes(axes[j])


fig.tight_layout()
plt.show()
```

# Explonatory Data Analysis

Univariate Analysis

```python
sns.histplot(data=df, x='pf_score', kde=True)
plt.title('Distribution of Personal Freedom Score')
plt.show()
```

```python
sns.histplot(data=df, x='hf_score', kde=True)
plt.title('Distribution of Human Freedom Score')
plt.show()
```

```python
sns.violinplot(data=df, y='ef_score')
plt.title('Violin Plot of Economic Freedom Score')
plt.show()
```

Bivariate analysis

```python
sns.scatterplot(data=df, x='ef_score', y='pf_score')
plt.title('Economic Freedom vs Personal Freedom')
plt.xlabel('Economic Freedom Score')
plt.ylabel('Personal Freedom Score')
plt.show()
```

```python

sns.lineplot(data=df, x='year', y='hf_score')
plt.title('Trend of Human Freedom Score Over Time')
plt.xlabel('Year')
plt.ylabel('Human Freedom Score')
plt.xticks(df['year'].unique()) 
plt.xticks(rotation=45)
plt.show()
```

```python
corr_matrix = df[['ef_score', 'pf_score', 'hf_score']].corr()
sns.heatmap(data=corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,8))
for index, group in df.groupby(['region']):
    ans = group.groupby(['year']).aggregate(np.mean)
    ans.plot(y='hf_score', label=index, ax=ax)
ax.legend(bbox_to_anchor=(1, 1))
plt.xticks(df['year'].unique()) 
plt.xticks(rotation=45)
plt.show()
```
